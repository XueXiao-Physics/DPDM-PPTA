```python
from numpy import array
import numpy as np
import matplotlib.pyplot as plt
filt=1873030
```

# U1B

```python
L=np.loadtxt('raw/raw_U1B')

lims=L[0:10]
mlhs=L[10:20]

lims[np.where(mlhs>filt)]=np.nan
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(3,1,figsize=(12,15))


mean=np.nanmean(lims,axis=0)
err=np.nanstd(lims,axis=0)/np.sqrt(10)
x=np.arange(0,125)/125*2.5-23.5+0.01
ax1.plot(x,lims.transpose(),'.');ax3.plot(x,mlhs.transpose(),'.');


nans=np.where(np.isnan(mean)==True)

mean=np.delete(mean,nans)
x=np.delete(x,nans)
err=np.delete(err,nans)


ax1.plot(x,mean,linewidth=4)
ax2.errorbar(x,mean,yerr=err,capsize=1.5,linewidth=2)



ax1.grid()
ax2.grid()
ax3.grid()

np.savetxt('U1B',(x,mean))
```
