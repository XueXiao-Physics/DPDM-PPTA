```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
ma=np.arange(0,125)/125.*2.5-23.5+0.01
a1=np.load('a1.npy')
a2=np.load('a2.npy')
a3=np.load('a3.npy')
a4=np.load('a4.npy')
b1=np.load('b1.npy')
b2=np.load('b2.npy')
b3=np.load('b3.npy')
b4=np.load('b4.npy')

ma=10**ma
a1=10**(2*a1)
a2=10**(2*a2)
a3=10**(2*a3)
a4=10**(2*a4)
b1=10**(2*b1)
b2=10**(2*b2)
b3=10**(2*b3)
b4=10**(2*b4)

range1=[10**-23.5,10**-21,10**(-52),10**(-41)]
range2=[1.5192e15*10**-23.5/2./np.pi,1.5192e15*10**-21/2./np.pi,10**(-52),10**(-41)]
```

```python


#plt.plot(data[0],data[2],label='Bayes Upper Limit')


fig,(fig1,fig2)=plt.subplots(1,2,figsize=(14,6))
plt.subplots_adjust(left=0.06,right=0.98)
#========================U(1)_{B}========================


fig1.set_xscale('log')
fig1.set_yscale('log')
fig1.set_xlabel(r'$U(1)_{B}$'+' dark photon mass (eV)')
fig1.set_ylabel(r'$\epsilon^2$ ' +'(dimensionless)')
fig1.axis(range1)


fig1.plot(ma[:len(a3)],a3,linewidth='1.6',label='PPTA_DR2_correlated(26-pulsars)',color='red')
fig1.plot(ma,a1,"--",label='PPTA_DR2_uncorrelated(26-pulsars)',color='red')

fig1.fill_between(ma[:len(a3)],a1[:len(a3)],a3,facecolor="yellow", alpha=0.5)

#MICROSCOPE
fig1.plot([10**-23.5,10**-18],[10**(-23.28*2),10**(-23.28*2)],'-',color='black' )
fig1.annotate('WEP experiments',horizontalalignment="left",xy=(10**-23.4,1.5*10**(-23.28*2)),xycoords='data')




fig1.fill_between([10**-23.5,10**-22],[10**-49.68,1e-52],facecolor="gray", alpha=0.1)
fig1.plot([10**-23.5,10**-22],[10**-49.68,1e-52],'--',color='black' )

fig1.legend(loc=2)
fig1.annotate(r"$U(1)_B$",horizontalalignment="right",bbox=dict(boxstyle="round",fc="w"),xy=(0.98,0.04),xycoords='axes fraction',size=15)


fig1x=fig1.twiny()
fig1x.set_xscale('log')
fig1x.set_yscale('log')
fig1x.axis(range2)
fig1x.set_xlabel('Frequency (Hz)')

#========================U(1)_{B-L}========================
fig2.set_xscale('log')
fig2.set_yscale('log')
fig2.set_xlabel(r'$U(1)_{B-L}$'+' dark photon mass (eV)')
fig2.set_ylabel(r'$\epsilon^2$ ' +'(dimensionless)')
fig2.axis(range1)

fig2.plot(ma[:len(a4)],a4,linewidth='1.6',color='red')
fig2.plot(ma,a2,"--",color='red')

fig2.fill_between(ma[:len(a4)],a2[:len(a4)],a4,facecolor="yellow", alpha=0.5)


fig2.plot([10**-23.5,10**-18],[10**(-24.07*2),10**(-24.07*2)],'-',color='black' )
fig2.annotate('WEP experiments',horizontalalignment="right",xy=(10**-21.1,1.5*10**(-24.07*2)),xycoords='data')

fig2.fill_between([10**-23.5,10**-22],[10**-49.68,1e-52],facecolor="gray", alpha=0.1)
fig2.plot([10**-23.5,10**-22],[10**-49.68,1e-52],'--',color='black' )

fig2.annotate(r"$U(1)_{B-L}$",horizontalalignment="right",bbox=dict(boxstyle="round",fc="w"),xy=(0.98,0.04),xycoords='axes fraction',size=15)


fig2x=fig2.twiny()
fig2x.set_xscale('log')
fig2x.set_yscale('log')
fig2x.axis(range2)
fig2x.set_xlabel('Frequency (Hz)')

plt.show()
#plt.savefig('upperlimit-logf.pdf')
```
