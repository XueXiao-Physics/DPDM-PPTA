```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig,(fig1,fig2)=plt.subplots(1,2,figsize=(12,5))
plt.subplots_adjust(left=0.06,right=0.98)

#========================U(1)_{B}========================
log10ma,lim=np.loadtxt('U1B.txt')
log10ma_corr,lim_corr=np.loadtxt('U1B_corr.txt')

ma=10**log10ma
f=1.51927e15*ma/2./np.pi
ma_corr=10**log10ma_corr
f_corr=1.51927e15*ma_corr/2./np.pi
lim=10**(2*lim)
lim_corr=10**(2*lim_corr)
range1=np.array([10**-23.49,10**-21.01,10**(-52),10**(-42)])
range2=np.array([10**-23.49*1.51927e15/2./np.pi,10**-21.01*1.51927e15/2./np.pi,10**(-52),10**(-42)])

fig1.set_xscale('log')
fig1.set_yscale('log')
fig1.set_xlabel(r'$U(1)_{B}$'+' dark photon mass (eV)',size=15)
fig1.set_ylabel(r'$\epsilon^2$ ' +'(dimensionless)',size=15)
fig1.tick_params(axis='x', labelsize=12 )
fig1.tick_params(axis='y', labelsize=12 )
fig1.axis(range1)


fig1.plot(ma_corr[:74],lim_corr[:74],linewidth='1.5',label='PPTA DR2 correlated(26 pulsars)',color='red')
fig1.plot(ma,lim,"--",linewidth='1.5',label='PPTA DR2 uncorrelated(26 pulsars)',color='red')


fig1.fill_between(ma,lim,1e-42,facecolor="gold", alpha=0.5)
fig1.fill_between(ma_corr[:74],lim_corr[:74],lim[:74],facecolor="yellow", alpha=0,hatch='//')

#MICROSCOPE
fig1.plot([10**-23.5,10**-18],[10**(-23.28*2),10**(-23.28*2)],'-',color='black' )
fig1.annotate('WEP experiments',horizontalalignment="left",xy=(10**-23.4,1.5*10**(-23.28*2)),xycoords='data')

#ultralight limit
#fig1.fill_between([10**-23.5,10**-22],[10**-42,10**-42],facecolor="steelblue", alpha=0.1)
#fig1.plot([10**-22,10**-22],[10**-52,10**-42])


fig1.fill_between([10**-23.5,10**-22],[10**-49.7,10**-52.7],facecolor="gray", alpha=0.3)
fig1.plot([10**-23.5,10**-22],[10**-49.7,10**-52.7],'--',color='black' )

fig1.legend(loc=2)
fig1.annotate(r"$U(1)_B$",horizontalalignment="right",bbox=dict(boxstyle="round",fc="w"),xy=(0.97,0.05),xycoords='axes fraction',size=20)


fig1x=fig1.twiny()
fig1x.set_xscale('log')
fig1x.set_yscale('log')
fig1x.axis(range2)
fig1x.set_xlabel('Frequency (Hz)',size=15)
fig1x.tick_params(axis='x', labelsize=12 )

#========================U(1)_{B-L}========================
log10ma,lim=np.loadtxt('U1BL.txt')
log10ma_corr,lim_corr=np.loadtxt('U1BL_corr.txt')

ma=10**log10ma
f=1.51927e15*ma/2./np.pi
ma_corr=10**log10ma_corr
f_corr=1.51927e15*ma_corr/2./np.pi
lim=10**(2*lim)
lim_corr=10**(2*lim_corr)
range1=np.array([10**-23.49,10**-21.01,10**(-52),10**(-42)])
range2=np.array([10**-23.49*1.51927e15/2./np.pi,10**-21.01*1.51927e15/2./np.pi,10**(-52),10**(-42)])

fig2.set_xscale('log')
fig2.set_yscale('log')
fig2.set_xlabel(r'$U(1)_{B-L}$'+' dark photon mass (eV)',size=15)
#fig2.set_ylabel(r'$\epsilon^2$ ' +'(dimensionless)',size=15)
fig2.tick_params(axis='x', labelsize=12 )
fig2.tick_params(axis='y', labelsize=12 )
fig2.axis(range1)

fig2.plot(ma_corr[:74],lim_corr[:74],linewidth='1.5',color='red',label='PPTA DR2 correlated(26 pulsars)')
fig2.plot(ma,lim,"--",linewidth='1.5',color='red',label='PPTA DR2 uncorrelated(26 pulsars)')

fig2.fill_between(ma,lim,1e-42,facecolor="gold", alpha=0.5)
fig2.fill_between(ma_corr[:74],lim_corr[:74],lim[:74],facecolor="yellow", alpha=0,hatch='//')

#MICROSCOPE
fig2.plot([10**-23.5,10**-18],[10**(-24.07*2),10**(-24.07*2)],'-',color='black' )
fig2.annotate('WEP experiments',horizontalalignment="right",xy=(10**-21.1,1.5*10**(-24.07*2)),xycoords='data')

fig2.fill_between([10**-23.5,10**-22],[10**-49.48,10**-52.48],facecolor="gray", alpha=0.3)
fig2.plot([10**-23.5,10**-22],[10**-49.48,10**-52.48],'--',color='black' )

fig2.legend(loc=2)
fig2.annotate(r"$U(1)_{B-L}$",horizontalalignment="right",bbox=dict(boxstyle="round",fc="w"),xy=(0.97,0.05),xycoords='axes fraction',size=20)


fig2x=fig2.twiny()
fig2x.set_xscale('log')
fig2x.set_yscale('log')
fig2x.axis(range2)
fig2x.set_xlabel('Frequency (Hz)',size=15)
fig2x.tick_params(axis='x', labelsize=12 )

fig.savefig('upperlimit.pdf',bbox_inches='tight')
```
